```python
# ===============================
# Import libraries
# ===============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# ===============================
# Load the dataset
# ===============================
df = pd.read_csv("NSF.CCME.csv")

# ===============================
# ===============================
# NSF Prediction
# ===============================
X_nsf = df[["NO3", "COD", "FC", "BOD", "TDS"]]
y_nsf = df["NSF"]

# Train-test split
X_train_nsf, X_test_nsf, y_train_nsf, y_test_nsf = train_test_split(
    X_nsf, y_nsf, test_size=0.3, random_state=42
)

# Scaling
scaler_nsf = StandardScaler()
X_train_nsf_scaled = scaler_nsf.fit_transform(X_train_nsf)
X_test_nsf_scaled = scaler_nsf.transform(X_test_nsf)

# MLP Model
mlp_nsf = MLPRegressor(
    hidden_layer_sizes=(3,),
    activation="logistic",
    alpha=0.0001,
    learning_rate="constant",
    solver="lbfgs",
    max_iter=10000,
    random_state=42
)
mlp_nsf.fit(X_train_nsf_scaled, y_train_nsf)
y_pred_nsf = mlp_nsf.predict(X_test_nsf_scaled)

# Metrics
r2_nsf = r2_score(y_test_nsf, y_pred_nsf)
rmse_nsf = np.sqrt(mean_squared_error(y_test_nsf, y_pred_nsf))
mae_nsf = mean_absolute_error(y_test_nsf, y_pred_nsf)

print("🔎 NSF Model Performance:")
print(f"R² Score : {r2_nsf:.4f}")
print(f"RMSE     : {rmse_nsf:.4f}")
print(f"MAE      : {mae_nsf:.4f}")

# Plot
plt.figure(figsize=(7,6))
plt.scatter(y_test_nsf, y_pred_nsf, color="blue", alpha=0.7, edgecolors="k", s=70)
lims = [38,50]
plt.plot(lims, lims, "r--", lw=2, label="1:1 Line")
plt.xlim(lims)
plt.ylim(lims)
plt.xticks(np.arange(38,52,2))
plt.yticks(np.arange(38,52,2))
plt.xlabel("Actual NSFWQI", fontsize=13, fontweight="bold")
plt.ylabel("Predicted NSFWQI", fontsize=13, fontweight="bold")
plt.legend()
plt.grid(False)
plt.savefig("NSF_WQI_Actual_vs_Predicted.png", dpi=1000, bbox_inches="tight")
plt.show()

# ===============================
# CCME Prediction
# ===============================
X_ccme = df[["COD", "FC", "BOD", "PO4", "NO3"]]
y_ccme = df["CCME"]

# Train-test split
X_train_ccme, X_test_ccme, y_train_ccme, y_test_ccme = train_test_split(
    X_ccme, y_ccme, test_size=0.3, random_state=42
)

# Scaling
scaler_ccme = StandardScaler()
X_train_ccme_scaled = scaler_ccme.fit_transform(X_train_ccme)
X_test_ccme_scaled = scaler_ccme.transform(X_test_ccme)

# MLP Model
mlp_ccme = MLPRegressor(
    hidden_layer_sizes=(3,),
    activation="logistic",
    alpha=0.0001,
    learning_rate="constant",
    solver="lbfgs",
    max_iter=10000,
    random_state=42
)
mlp_ccme.fit(X_train_ccme_scaled, y_train_ccme)
y_pred_ccme = mlp_ccme.predict(X_test_ccme_scaled)

# Metrics
r2_ccme = r2_score(y_test_ccme, y_pred_ccme)
rmse_ccme = np.sqrt(mean_squared_error(y_test_ccme, y_pred_ccme))
mae_ccme = mean_absolute_error(y_test_ccme, y_pred_ccme)

print("🔎 CCME Model Performance:")
print(f"R² Score : {r2_ccme:.4f}")
print(f"RMSE     : {rmse_ccme:.4f}")
print(f"MAE      : {mae_ccme:.4f}")

# Plot
plt.figure(figsize=(7,6))
plt.scatter(y_test_ccme, y_pred_ccme, color="green", alpha=0.7, edgecolors="k", s=70)
lims_ccme = [20,100]  # Adjust based on your CCME range
plt.plot(lims_ccme, lims_ccme, "r--", lw=2, label="1:1 Line")
plt.xlim(lims_ccme)
plt.ylim(lims_ccme)
plt.xlabel("Actual CCME WQI", fontsize=13, fontweight="bold")
plt.ylabel("Predicted CCME WQI", fontsize=13, fontweight="bold")
plt.legend()
plt.grid(False)
plt.savefig("CCME_WQI_Actual_vs_Predicted.png", dpi=1000, bbox_inches="tight")
plt.show()
```

```python
# mlp_models.py
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
import joblib

# Load dataset
df = pd.read_csv("NSF.CCME.csv")

# ================= NSF MODEL =================
X_nsf = df[["NO3", "COD", "FC", "BOD", "TDS"]]
y_nsf = df["NSF"]

X_train_nsf, X_test_nsf, y_train_nsf, y_test_nsf = train_test_split(X_nsf, y_nsf, test_size=0.3, random_state=42)

scaler_nsf = StandardScaler()
X_train_nsf_scaled = scaler_nsf.fit_transform(X_train_nsf)
X_test_nsf_scaled = scaler_nsf.transform(X_test_nsf)

mlp_nsf = MLPRegressor(hidden_layer_sizes=(3,), activation="logistic",
                       alpha=0.0001, learning_rate="constant", solver="lbfgs",
                       max_iter=10000, random_state=42)
mlp_nsf.fit(X_train_nsf_scaled, y_train_nsf)

# Save NSF model and scaler
joblib.dump(mlp_nsf, "mlp_nsf_model.pkl")
joblib.dump(scaler_nsf, "scaler_nsf.pkl")

# ================= CCME MODEL =================
X_ccme = df[["COD", "FC", "BOD", "PO4", "NO3"]]
y_ccme = df["CCME"]

X_train_ccme, X_test_ccme, y_train_ccme, y_test_ccme = train_test_split(X_ccme, y_ccme, test_size=0.3, random_state=42)

scaler_ccme = StandardScaler()
X_train_ccme_scaled = scaler_ccme.fit_transform(X_train_ccme)
X_test_ccme_scaled = scaler_ccme.transform(X_test_ccme)

mlp_ccme = MLPRegressor(hidden_layer_sizes=(3,), activation="logistic",
                        alpha=0.0001, learning_rate="constant", solver="lbfgs",
                        max_iter=10000, random_state=42)
mlp_ccme.fit(X_train_ccme_scaled, y_train_ccme)

# Save CCME model and scaler
joblib.dump(mlp_ccme, "mlp_ccme_model.pkl")
joblib.dump(scaler_ccme, "scaler_ccme.pkl")

print("✅ Models saved successfully!")
```

```python
# app.py
import streamlit as st
import numpy as np
import joblib

st.set_page_config(page_title="Water Quality Index Predictor", layout="centered")

st.title("🌊 Water Quality Index Predictor")
st.markdown("Predict **NSF WQI** and **CCME WQI** based on water quality parameters.")

# Load models and scalers
mlp_nsf = joblib.load("mlp_nsf_model.pkl")
scaler_nsf = joblib.load("scaler_nsf.pkl")
mlp_ccme = joblib.load("mlp_ccme_model.pkl")
scaler_ccme = joblib.load("scaler_ccme.pkl")

st.header("NSF WQI Input Parameters")
NO3 = st.number_input("Nitrate (NO₃⁻)", min_value=0.0, max_value=100.0, value=10.0)
COD = st.number_input("COD (mg/L)", min_value=0.0, max_value=500.0, value=30.0)
FC = st.number_input("Fecal Coliform (FC)", min_value=0.0, max_value=1000.0, value=50.0)
BOD = st.number_input("BOD (mg/L)", min_value=0.0, max_value=50.0, value=5.0)
TDS = st.number_input("TDS (mg/L)", min_value=0.0, max_value=2000.0, value=300.0)

st.header("CCME WQI Input Parameters")
PO4 = st.number_input("Phosphate (PO₄³⁻)", min_value=0.0, max_value=20.0, value=2.0)

if st.button("Calculate WQIs"):
    # Prepare inputs
    X_nsf_input = np.array([[NO3, COD, FC, BOD, TDS]])
    X_nsf_scaled = scaler_nsf.transform(X_nsf_input)
    nsf_pred = mlp_nsf.predict(X_nsf_scaled)[0]

    X_ccme_input = np.array([[COD, FC, BOD, PO4, NO3]])
    X_ccme_scaled = scaler_ccme.transform(X_ccme_input)
    ccme_pred = mlp_ccme.predict(X_ccme_scaled)[0]

    st.subheader("💧 Predicted Water Quality Indexes")
    st.write(f"**NSF WQI:** {nsf_pred:.2f}")
    st.write(f"**CCME WQI:** {ccme_pred:.2f}")

    # Interpretation (example)
    def nsf_interpret(wqi):
        if wqi > 90:
            return "Excellent"
        elif wqi > 70:
            return "Good"
        elif wqi > 50:
            return "Fair"
        else:
            return "Poor"

    def ccme_interpret(wqi):
        if wqi > 80:
            return "Excellent"
        elif wqi > 60:
            return "Good"
        elif wqi > 45:
            return "Fair"
        else:
            return "Poor"

    st.write(f"**NSF WQI Interpretation:** {nsf_interpret(nsf_pred)}")
    st.write(f"**CCME WQI Interpretation:** {ccme_interpret(ccme_pred)}")
```
